# Part 1

```python
sum = 0
for (begin, end) in data:
    for i in range(begin, end + 1):
        word = str(i)
        size = len(word)
        if size % 2 == 0:
            if word[:size//2] == word[size//2:]:
                sum += i
sum
```

# Part 2

```python
import re
regex = r'^([0-9]+)\1{1,}$'
sum = 0
for (begin, end) in data:
    for i in range(begin, end + 1):
        word = str(i)
        if re.match(regex, word):
            sum += i
sum
```

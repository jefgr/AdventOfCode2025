# Part 2

```python
from itertools import combinations, pairwise

v_edges = []
h_edges = []

prev_y, prev_x = data[-1]

class vedge():
    def __init__(self, x, y_min, y_max):
        self.x = x
        self.y_min = y_min
        self.y_max = y_max

class hedge():
    def __init__(self, y, x_min, x_max):
        self.y = y
        self.x_min = x_min
        self.x_max = x_max

prev_x, prev_y = data[-1][0], data[-1][1]
for (y, x) in data:
    if x == prev_x:
        y_min, y_max = min(y, prev_y), max(y, prev_y)
        v_edges.append(vedge(x, y_min, y_max))
    if y == prev_y:
        x_min, x_max = min(x, prev_x), max(x, prev_x)
        h_edges.append(hedge(y, x_min, x_max))
    prev_x, prev_y = x, y

def check_edges_for_collision(p1, p2):
    y1, x1 = p1
    y2, x2 = p2

    min_x, max_x = min(x1, x2), max(x1, x2)
    min_y, max_y = min(y1, y2), max(y1, y2)

    for edge in v_edges:
        if min_x < edge.x < max_x:
            if max(edge.y_min, min_y) < min(edge.y_max, max_y):
                return True
            
    for edge in h_edges:
        if min_y < edge.y < max_y:
            if max(edge.x_min, min_x) < min(edge.x_max, max_x):
                return True
    return False


def size(a, b):
    x = abs(a[0] - b[0]) + 1
    y = abs(a[1] - b[1]) + 1
    return x * y

curr_s = 0

for combo in combinations(data, 2):
    p1, p2 = combo
    if size(p1, p2) <= curr_s:
        continue
    if check_edges_for_collision(p1, p2):
        continue
    curr_s = size(p1, p2)


curr_s
```

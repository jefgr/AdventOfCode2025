```python
import copy

def rotate(shape):
    rotated = list(zip(*(shape[::-1])))
    return rotated


def all_rotations(shape):
    rotations = [shape]
    current = shape
    for _ in range(3):
        current = rotate(current)
        rotations.append(current)
    return rotations


def can_place_present(grid, present, row, col):
    rows = len(present)
    cols = len(present[0])
    
    if row + rows > len(grid) or col + cols > len(grid[0]):
        return False
    
    for i in range(rows):
        for j in range(cols):
            if present[i][j] == 1 and grid[row + i][col + j] == 1:
                return False
    
    return True


def place_present(grid, present, row, col):
    new_grid = copy.deepcopy(grid)
    for i, row in enumerate(present):
        for j, val in enumerate(row):
            if val == 1:
                new_grid[row + i][col + j] = 1
    return new_grid


def try_pack_presents(width, height, present_list):
    grid = [[0] * width for _ in range(height)]
    def backtrack(index):
        if index == len(present_list):
            return True
        
        present_type, count = present_list[index]
        shape = presents[present_type]
        rotations = all_rotations(shape)
        
        for _ in range(count):
            placed = False
            for rotation in rotations:
                if placed:
                    break
                for row in range(height):
                    if placed:
                        break
                    for col in range(width):
                        if can_place_present(grid, rotation, row, col):
                            place_present_in_place(grid, rotation, row, col)
                            placed = True
                            break
            if not placed:
                return False
            return backtrack(index + 1)
    
    def place_present_in_place(grid, present, row, col):
        for i in range(len(present)):
            for j in range(len(present[0])):
                if present[i][j] == 1:
                    grid[row + i][col + j] = 1
    
    return backtrack(0)

total = 0

def check_all_trees():
    global total    
    for tree in trees_data:
        width = tree[0][0]
        height = tree[0][1]
        
        present_counts = tree[1]
        present_list = [(i, present_counts[i]) for i in range(len(present_counts))]

        total_area_needed = sum(
            sum(sum(row) for row in presents[ptype]) * count 
            for ptype, count in present_list
        )
        tree_area = width * height
        
        if total_area_needed <= tree_area:
            fits = try_pack_presents(width, height, present_list)
            total += 1

check_all_trees()
total
```

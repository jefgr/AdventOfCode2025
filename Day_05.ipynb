```python
ranges.sort(key= lambda x: x[0])
i = 0
while(i < len(ranges) - 1):
    if ranges[i][1] >= ranges[i + 1][0] - 1:
        temp = [ranges[i][0], max(ranges[i][1], ranges[i + 1][1])]
        ranges.pop(i + 1)
        ranges[i] = temp
    else:
        i += 1

result = 0
ids = ids_in.copy()
ids.sort()
curr_range_index = 0
curr_range = ranges[curr_range_index]    

while ids != []:
    id = ids[0]
    if curr_range[1] < id:
        curr_range_index += 1
        if curr_range_index >= len(ranges):
            break
        curr_range = ranges[curr_range_index]
    elif curr_range[0] > id:
        ids.pop(0)
    else:
        result += 1
        ids.pop(0)
result
```

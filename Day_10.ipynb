```python
data = [line.split(']') for line in input.splitlines()]
lights = [[True if a == '#' else False for a in line[0][1:]] for line in data]
inter = [line[1].split('{') for line in data]
buttons = [[tuple(int(c) for c in a.replace('(', '').replace(')', '').split(',')) for a in line[0].strip().split(" ")] for line in inter]
joltages = [[int(a) for a in line[1][:-1].split(',')] for line in inter]
lights
```

```python
class row():
    def __init__(self, lights, buttons, joltages):
        self.lights = lights
        self.buttons = buttons
        self.joltages = joltages
    def toString(self):
        return f"{self.lights} ; {self.buttons} ; {self.joltages}"

rows = [row(a, b, c) for a, b, c in zip(lights, buttons, joltages)]
rows[0].toString()
```

# Part 1

```python
from itertools import product
def find_button_presses(row: row):
    i = 1
    b = row.buttons
    l = row.lights
    curr_l = [False for a in l]
    not_found = True
    while (not_found):
        for combo in product(b, repeat=i):
            temp_l = curr_l.copy()
            for el in combo:
                for n in el:
                    temp_l[n] = not temp_l[n]
            if temp_l == l:
                not_found = False
                return i
        i += 1

sum(find_button_presses(r) for r in rows)
```

# Part 2

```python
from itertools import product
from z3 import Int, Optimize, sat

def find_button_presses_for_joltages(row: row):
    i = 1
    b = row.buttons
    l = row.joltages
    solver = Optimize()
    variables = []
    joltages_list = [None] * len(l)

    for i, button in enumerate(b):
        var = Int(str(i))
        variables.append(var)
        solver.add(var >= 0)
        
        for x in button:
            if joltages_list[x] is None:
                joltages_list[x] = var
            else:
                joltages_list[x] = joltages_list[x] + var
    
    for i, jolt in enumerate(l):
        if joltages_list[i] is None:
            continue
        solver.add(jolt == joltages_list[i])
    
    presses = solver.minimize(sum(variables))

    if solver.check() == sat:
            return presses.value().as_long()
    else:
        assert False



s = 0
for r in rows:
    temp = find_button_presses_for_joltages(r)
    s += temp
s
```

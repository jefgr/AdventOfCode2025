```python
data = input.splitlines()
data
```

# Part 1

```python
beams = {data[0].index('S')}
lines = [c for c in data if '^' in c]
split_amount = 0
for line in lines:
    new_beams = set()
    for beam in beams:
        if line[beam] == '^':
            split_amount += 1
            if beam - 1 >= 0:
                new_beams.add(beam - 1)
            if beam + 1 < len(line):
                new_beams.add(beam + 1)
        else:
            new_beams.add(beam)
    beams = new_beams
split_amount
```

# Part 2

```python
matrix = [[1 if '^' == c else 0 for c in line] for line in data if '^' in line or 'S' in line]
matrix
```

```python
start = data[0].index('S')
matrix2 = [[0 for _ in line] for line in matrix]
matrix2[0][start] = 1
for heigth, line in enumerate(matrix2):
    for width, n in enumerate(line):
        if matrix[heigth][width]:
            if width > 0:
                matrix2[heigth][width - 1] += matrix2[heigth - 1][width]
            if width + 1 < len(matrix2[heigth]):
                matrix2[heigth][width + 1] += matrix2[heigth - 1][width]
        else:
            matrix2[heigth][width] += matrix2[heigth - 1][width]
sum(matrix2[-1])
```

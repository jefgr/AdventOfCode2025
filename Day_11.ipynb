# Part 2

```python
from functools import cache

@cache
def dft(dac, fft, name):
    amount_of_paths = 0
    if name == 'out':
        if dac and fft:
            return 1
        return 0
    for node in d[name]:
        if name == 'dac':
            amount_of_paths += dft(True, fft, node)
        elif name == 'fft':
            amount_of_paths += dft(dac, True, node)
        else:
            amount_of_paths += dft(dac, fft, node)
    return amount_of_paths

dft(False, False, 'svr')
```

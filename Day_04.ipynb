```python
import numpy as np
from scipy.signal import convolve2d
npdata = np.array(data)

mask = np.array([[1, 1, 1],[1, 0, 1],[1, 1, 1]])
sumsarray = convolve2d(npdata, mask, boundary='fill', fillvalue=0, mode='same')

resultmatrix = np.multiply(npdata, sumsarray)
boolmatrix = np.multiply((resultmatrix < 4), npdata)
np.count_nonzero(boolmatrix)
```

# Part 2

```python
npdata = np.array(data)
test = True
mask = np.array([[1, 1, 1],[1, 0, 1],[1, 1, 1]])
sum = 0
while(test):
    test = False
    sumsarray = convolve2d(npdata, mask, boundary='fill', fillvalue=0, mode='same')
    resultmatrix = np.multiply(npdata, sumsarray)
    boolmatrix = np.multiply((resultmatrix < 4), npdata)
    amount_removed = np.count_nonzero(boolmatrix)
    if amount_removed > 0:
        test = True
        sum += amount_removed
        npdata = np.bitwise_xor(npdata, boolmatrix)
sum
```

```python
%matplotlib inline
from matplotlib import rc
rc('animation', html='jshtml')

import numpy as np
from scipy.signal import convolve2d
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# ---------------------------
# Your starting data
# ---------------------------
npdata = np.array(data)

mask = np.array([[1, 1, 1],
                 [1, 0, 1],
                 [1, 1, 1]])

frames = []   # will store (npdata, boolmatrix)


# ---------------------------
# Generate frames
# ---------------------------
test = True
while test:
    test = False
    sumsarray = convolve2d(npdata, mask, boundary='fill', fillvalue=0, mode='same')
    resultmatrix = np.multiply(npdata, sumsarray)
    boolmatrix = np.multiply((resultmatrix < 4), npdata)

    frames.append((npdata.copy(), boolmatrix.copy()))

    if np.count_nonzero(boolmatrix) > 0:
        test = True
        npdata = np.bitwise_xor(npdata, boolmatrix)


# ---------------------------
# Setup figure (persistent objects)
# ---------------------------
fig, ax = plt.subplots(figsize=(6, 6))

np_img = ax.imshow(frames[0][0], cmap="gray_r", vmin=0, vmax=1)

# highlight layer using masked array (best for Jupyter)
highlight = np.ma.masked_where(frames[0][1] == 0, frames[0][1])
hl_img = ax.imshow(highlight, cmap="Reds", vmin=0, vmax=1)

ax.set_xticks([])
ax.set_yticks([])


# ---------------------------
# Update function
# ---------------------------
def update(i):
    np_frame, bool_frame = frames[i]

    np_img.set_data(np_frame)

    highlight = np.ma.masked_where(bool_frame == 0, bool_frame)
    hl_img.set_data(highlight)

    ax.set_title(f"Iteration {i}")
    return [np_img, hl_img]


# ---------------------------
# Create looping animation
# ---------------------------
ani = FuncAnimation(
    fig,
    update,
    frames=len(frames),
    interval=600,
    repeat=True
)

ani  # <-- In Jupyter, simply returning ani SHOWS the animation
```
